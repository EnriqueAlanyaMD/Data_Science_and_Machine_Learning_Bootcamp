# Example 1: A Simple Cost Function

$f(x) = x^2 + x + 1$

```python
def f(x):
    return x**2 + x + 1
```

```python
# Make data
x_1 = np.linspace(start=-3, stop=3, num=100)
x_1
```

```python
# Plot
plt.xlim(-3,3)
plt.ylim(0, 8)
plt.xlabel("X", fontsize=16)
plt.ylabel("f(x)", fontsize=16)

plt.plot(x_1, f(x_1))
plt.show()
```

# Understanding the Power Rule & Creating Charts with Subplots

## Slope & Derivatives

```python
# Create a Python function for derivative of f(x) called df(x)

def df(x):
    return 2*x + 1
```

```python
# Plot function and derivative side by side
plt.figure(figsize=[20,5])

# 1. Function
plt.subplot(1,2,1)
plt.title('Cost Function')
plt.xlim(-3,3)
plt.ylim(0, 8)
plt.xlabel("X", fontsize=16)
plt.ylabel("f(x)", fontsize=16)
plt.plot(x_1, f(x_1), color='blue', linewidth=5)

# 2. Derivative
plt.subplot(1,2,2)
plt.title('Slope of the Cost Function')
plt.xlim(-1,2)
plt.ylim(0, 6)
plt.xlabel("X", fontsize=16)
plt.ylabel("df(x)", fontsize=16)
plt.grid()
plt.plot(x_1, df(x_1), color='skyblue', linewidth=5)

plt.show()
```

# Python Loops and the Gradient Descent Algorithm

```python
# Python For Loop
for n in range(5):
    print(n, 'Hello World')
```

```python
# Python while loop
counter = 0
while counter < 7:
    print('Counting ...', counter)
    counter = counter + 1
print('Ready or not, here I come!')
```

```python
# Gradient Descent

## Starting Point
new_x = 3
## Temp Value
prev_x = 0
## Learning Rate (gamma)
step_multiplier = 0.1

precision = 0.0001

x_list = [new_x]
slope_list = [df(new_x)]

for n in range(100):
    # Step 1 - Make a guess/prediction
    prev_x = new_x
    # Step 2 - Calculate the error 
    gradient = df(prev_x)
    # Step 3 - Learning
    new_x = prev_x - step_multiplier * gradient
    
    step_size = abs(new_x - prev_x)
    
    x_list.append(new_x)
    slope_list.append(df(new_x))
    
    if step_size < precision:
        print('Meets precision at:', n)
        break
    
print('Local minimum occurs at: ', new_x)
print('Slope or df(x) value at this point is:', df(new_x))
print('f(x) value or cost at this point is', f(new_x))
```

```python
# Superimpose the gradient descent calculations on plot
plt.figure(figsize=[20,5])

# 1. Function
plt.subplot(1,3,1)
plt.title('Cost Function')
plt.xlim(-3,3)
plt.ylim(0, 8)
plt.xlabel("X", fontsize=16)
plt.ylabel("f(x)", fontsize=16)
plt.plot(x_1, f(x_1), color='blue', linewidth=5, alpha=0.7)

# Convert to an array
values = np.array(x_list)
plt.scatter(x_list, f(values), color='red',s=100, alpha=0.6)

# 2. Derivative
plt.subplot(1,3,2)
plt.title('Slope of the Cost Function')
plt.xlim(-1,2)
plt.ylim(0, 6)
plt.xlabel("X", fontsize=16)
plt.ylabel("df(x)", fontsize=16)
plt.grid()
plt.plot(x_1, df(x_1), color='skyblue', linewidth=5, alpha=0.7)

plt.scatter(x_list, slope_list, color='red', s = 100, alpha=0.6)

# 3. Derivative (Close Up)
plt.subplot(1,3,3)
plt.title('Gradient Descent (close up)')
plt.xlim(-0.55, -0.2)
plt.ylim(-0.5, 0.8)
plt.xlabel("X", fontsize=16)
plt.ylabel("df(x)", fontsize=16)
plt.grid()
plt.plot(x_1, df(x_1), color='skyblue', linewidth=5, alpha=0.7)

plt.scatter(x_list, slope_list, color='red', s = 300, alpha=0.6)

plt.show()
```
