```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

# Example 1: A Simple Cost Function

$f(x) = x^2 + x + 1$

```python
def f(x):
    return x**2 + x + 1
```

```python
# Make data
x_1 = np.linspace(start=-3, stop=3, num=100)
x_1
```

```python
# Plot
plt.xlim(-3,3)
plt.ylim(0, 8)
plt.xlabel("X", fontsize=16)
plt.ylabel("f(x)", fontsize=16)

plt.plot(x_1, f(x_1))
plt.show()
```

# Understanding the Power Rule & Creating Charts with Subplots

## Slope & Derivatives

```python
# Create a Python function for derivative of f(x) called df(x)

def df(x):
    return 2*x + 1
```

```python
# Plot function and derivative side by side
plt.figure(figsize=[20,5])

# 1. Function
plt.subplot(1,2,1)
plt.title('Cost Function')
plt.xlim(-3,3)
plt.ylim(0, 8)
plt.xlabel("X", fontsize=16)
plt.ylabel("f(x)", fontsize=16)
plt.plot(x_1, f(x_1), color='blue', linewidth=5)

# 2. Derivative
plt.subplot(1,2,2)
plt.title('Slope of the Cost Function')
plt.xlim(-1,2)
plt.ylim(0, 6)
plt.xlabel("X", fontsize=16)
plt.ylabel("df(x)", fontsize=16)
plt.grid()
plt.plot(x_1, df(x_1), color='skyblue', linewidth=5)

plt.show()
```
